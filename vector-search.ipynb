```python
import pandas as pd
books = pd.read_csv("books_cleaned.csv")
```

```python
books["tagged_description"].to_csv("tagged_description.txt",sep ="\n", index = False,header = False)
```

```python
raw_documents= TextLoader("tagged_description.txt",encoding="utf-8").load()
text = raw_documents[0].page_content
matches = re.findall(r'(978\d{10})\s+(.*?)(?=(978\d{10})|$)', text, flags=re.DOTALL)
documents = [
    Document(page_content=desc.strip(), metadata={"isbn": isbn})
    for isbn, desc, _ in matches
]
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
db_books = Chroma(
    persist_directory="./book_vectors",  # path to your folder
    embedding_function=embeddings
)
```

```python
query = "A book to teach children about nature"
docs = db_books.similarity_search(query, k=10)
docs
```

```python
isbn = int(docs[0].metadata["isbn"])
matching_book = books[books["isbn13"] == isbn]
matching_book
```

```python
def retrieve_semantic_recommendations(
    query: str,
    top_k: int = 10,
) -> pd.DataFrame:
    # Run similarity search
    recs = db_books.similarity_search(query, k=50)

    # Extract ISBNs safely
    books_list = []
    for doc in recs:
        if "isbn" in doc.metadata:
            try:
                books_list.append(int(doc.metadata["isbn"]))
            except ValueError:
                # skip if ISBN is not a valid number
                continue

    # Ensure same type in DataFrame
    books_copy = books.copy()
    books_copy["isbn13"] = books_copy["isbn13"].astype(int)

    # Filter matching books
    matching_books = books_copy[books_copy["isbn13"].isin(books_list)]

    # Return top_k
    return matching_books.head(top_k)
```

```python
retrieve_semantic_recommendations("A book to teach children about nature")
```
